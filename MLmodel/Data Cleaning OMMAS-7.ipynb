```python
import re
from fpdf import FPDF

def process_text_report(text_report):
    # Split the text into individual items
    items = re.split(r'\d+\..', text_report)
    items = [item.strip() for item in items if item.strip()]

    # Create and return a dictionary of discrepancies and measures
    discrepancy_to_measure = {}
    for item in items:
        item = ' '.join(item.split())  # Clean up the text
        discrepancy_and_measure = item.split(':', 1)
        if len(discrepancy_and_measure) == 2:
            discrepancy, measure = discrepancy_and_measure
            discrepancy_to_measure[discrepancy.strip()] = measure.strip()
    return discrepancy_to_measure

class PDF(FPDF):
    def add_title(self, title):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, title, 0, 1, 'C')

    def add_discrepancy(self, discrepancy, measure):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, f"Discrepancy: {discrepancy}", 0, 1)
        self.set_font('Arial', '', 12)
        self.multi_cell(0, 10, f"Corrective Measure: {measure}\n")

# Your text report
text_report = """
1._Maintenance of Road :

Rain Cuts: Restore using proper backfilling and compaction. Use geotextiles if necessary for stabilization.
Vegetation: Regularly trim branches, shrubs, and grass. Ensure clear visibility for road users and prevent obstruction of traffic signs.

2._Making up of Earthen Shoulders:

Ensure proper grading and compaction to match the road level. Regular maintenance is needed to prevent erosion.

3._Condition of Pavement :

Potholes: Fill with appropriate materials (asphalt or concrete) followed by compaction.
Patch Repairs: Use a cut-and-patch method for larger areas, ensuring the patch is well integrated with the existing pavement.

4._Road Furniture and Markings:

Regularly repaint road markings.
Maintain visibility and positioning of road furniture.

5._Geometrics:

Regularly assess and maintain road alignment, curvature, and gradients to meet design standards.

6._Base Course - Water Bound Macadam (WBM):

Regularly inspect for surface evenness, and presence of potholes or ruts.
For repairs, remove the loose material, re-grade and re-compact the area.

7._Cross drainage works:

Regular cleaning to prevent blockages.
Structural inspection and repair of any damages to maintain functionality.
In case of significant damage, reconstruct the affected area with fresh WBM layers, ensuring proper gradation of aggregates and adequate binding with stone dust or fine aggregates.
Ensure proper water drainage to avoid water logging which can weaken the WBM structure.

"""

# Process the text report
discrepancy_to_measure = process_text_report(text_report)

# Create a PDF object
pdf = PDF()
pdf.add_page()

# Add a title
pdf.add_title('Corrective Measures Report')

# Add each discrepancy and its corrective measure to the PDF
for discrepancy, measure in discrepancy_to_measure.items():
    pdf.add_discrepancy(discrepancy, measure)

# Save the PDF
pdf.output("Corrective Measures .pdf")
```
